```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

**data generation**
y=x^2+5*x+3

```python
x=np.arange(-80,81,.5)
y=x**2+5*x+3
```

```python
x.shape,y.shape
```

**train/test**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

**model definition**

```python
#model1
model1=keras.Sequential()
model1.add(keras.layers.Dense(units=1024,activation='relu',input_shape=[1]))
model1.add(keras.layers.Dense(units=1))
```

```python
model1.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model1.summary()
```

```python
hist_1=model1.fit(x_train,y_train,batch_size=64, epochs=1000,validation_data=(x_test,y_test))
```

```python
#model2
model2=keras.Sequential()
model2.add(keras.layers.Dense(units=256,activation='relu', input_shape=[1]))
model2.add(keras.layers.Dense(units=256,activation='relu'))
model2.add(keras.layers.Dense(units=256,activation='relu'))
model2.add(keras.layers.Dense(units=256,activation='relu')) #these are hidden layers
model2.add(keras.layers.Dense(units=1))  #output layer
```

```python
model2.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model2.summary()
```

```python
model2.fit(x_train,y_train,batch_size=64, epochs=1000,validation_data=(x_test,y_test))
```

**comparision**

```python
y_pred1=model1.predict(x_test)
y_pred2=model2.predict(x_test)
```

```python
plt.scatter(x_test,y_test,color='red',linewidths=4)
plt.scatter(x_test,y_pred1,color='pink')
plt.scatter(x_test,y_pred2,color='purple')
plt.show()
```
